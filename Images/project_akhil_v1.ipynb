```python
#Morgage calculator
def morgagecalculator (morgage, years, interest):
    interest = interest / 100
    nper = years*12
    interest_monthly = interest/12
    numerator = interest_monthly*((1+interest_monthly)**nper)
    denominator = (1+interest_monthly)**nper-1
    payment = float("{0:.2f}".format(morgage*numerator/denominator))
    return(payment)
```

```python
x = int(input('Please enter your mortgage value'))
y = int(input('Please enter the number of years'))
z = int(input('Please enter the intrest rate'))
morgagecalculator(x,y,z)
```
