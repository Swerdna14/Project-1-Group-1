# Looking for the house in Perth
It is much easier to find the house in the the new city if you have proper tool that can help you to direct the search.
We offer the application that can help the new buyer to locate the suburbs where he/she can find the property sutisfying initial parametrs of the buyer.

```python
#Importing libraries
import pandas as pd
from pathlib import Path
import os
import hvplot.pandas
import numpy as np
from datetime import date
```

```python
#Creating path
project_path = Path("./all_perth_310121.csv")
```

```python
#Reading csv file
df = pd.read_csv(project_path)
df.head()
```

```python
def parse_date(date): 
    date = date.split("-")[-1] 
    date = date.replace("\r","") 
    return int(date)
```

```python
df['DATE_SOLD'] = df['DATE_SOLD'].apply(parse_date)
#df
```

```python
df['DATE_SOLD'].value_counts().sort_index()
```

```python
UPDATING_DICTIONARY = {2014: 5.6, 
                       2015: 8.9, 
                       2016: 13.5,
                       2017: 15.3,
                       2018: 17.9,
                       2019: 20.5,
                       2020: 16.3, 
                       2021: 6}
```

```python
df = df[ df['DATE_SOLD'].isin(UPDATING_DICTIONARY) ].copy() # Here we remove all the elements not in the updating dict
```

```python
df.columns
```

```python
def update_value(row):
    date_sold = row['DATE_SOLD']
    multiplier = UPDATING_DICTIONARY[date_sold]
    price = row['PRICE']
    updated_price = price * (1 + multiplier / 100)
    return updated_price
```

```python
df['UPDATED_PRICE'] = df.apply(update_value, axis = 1)
```

In the beginning of the process the buyer could observe the map of the Perth with all the suburbs and color points reflecting price renge of the houses.

```python
#Creating a new column for Price_PSM
df['PRICE_PSM'] = df['UPDATED_PRICE'] / df['FLOOR_AREA']
df['PRICE_PSM'] =  round(df['PRICE_PSM'], 0)
df.head()
```

```python
#Creating representative DataFrame 
price_per_meter = df[['SUBURB', 'PRICE_PSM', 'LATITUDE', 'LONGITUDE']].groupby(['SUBURB']).mean()
price_per_meter = price_per_meter.reset_index()
price_per_meter.head()
```

```python
# Creating the plot 
price_per_meter.hvplot.points(
    'LONGITUDE', 
    'LATITUDE', 
    geo=True, 
    size='PRICE_PSM',
    scale=.1,
    color='PRICE_PSM',
    tiles='OSM',
    frame_width=700,
    frame_height=500
    )
```

If the initial parametrs of the buyer include number of bedrums and and specific suburb, we can provide the information about the potential budget 

```python
#Creating DataFrame for the BUDGET calculation
price_by_dist_bed = df[['SUBURB', 'BEDROOMS', 'UPDATED_PRICE']].groupby(['SUBURB','BEDROOMS']).mean()
price_by_dist_bed = price_by_dist_bed.reset_index()
price_by_dist_bed['UPDATED_PRICE'] = round(price_by_dist_bed['UPDATED_PRICE'], 0)
price_by_dist_bed
```

```python
#Budget calculator
#Inputs
bedrums = int(input('How many rooms do you want to have in your house? '))
suburb = input('In what suburb? ') 
budget = price_by_dist_bed.loc[(price_by_dist_bed['BEDROOMS'] == bedrums) & (price_by_dist_bed['SUBURB'] == suburb)]
budget
```

In case when the initial parametrs of the buyer include number of bedrums and specific badget, we can provide the information about the potential suburbs

```python
#Districts 
plan_bedrums = int(input('How many rooms do you want to have in your house? '))
plan_budget = int(input('What would be your budget? '))
plan_suburbs = price_by_dist_bed.loc[(price_by_dist_bed['BEDROOMS'] == plan_bedrums) & (price_by_dist_bed['PRICE'] <= plan_budget)]
plan_suburbs
```

The application can also provide the calculation of the monthly payment in case if the buyer consider using the credit of the bank.

```python
#Morgage calculator
def mortgagecalculator (mortgage, years, interest):
    interest = interest / 100
    nper = years*12
    interest_monthly = interest/12
    numerator = interest_monthly*((1+interest_monthly)**nper)
    denominator = (1+interest_monthly)**nper-1
    payment = float("{0:.2f}".format(mortgage*numerator/denominator))
    return(payment)
```

```python
x=int(input('What is the mortgage amount? '))
y=int(input('For how many years? '))
z=int(input('What interest rate? '))
mortgagecalculator(x,y,z)
```

As well as stamp duti calculation and transfer registration fee

```python
def stampdutycalculator (purchase_price):
    if purchase_price <= 120000:
        stamp_duty = purchase_price*0.019
    elif purchase_price > 120000 and purchase_price <= 150000:
        stamp_duty = 2280 + (purchase_price - 120000) * 0.0285
    elif purchase_price > 150000 and purchase_price <= 360000:
        stamp_duty = 3135 + (purchase_price - 150000) * 0.038
    elif purchase_price > 360000 and purchase_price <= 725000:
        stamp_duty = 11175 + (purchase_price - 360000) * 0.0475
    else:
        stamp_duty = 28453 + (purchase_price - 725000) * 0.0515
    
    return stamp_duty
    
```

```python
q=int(input('What is the property value? '))
stampdutycalculator(q)
```
